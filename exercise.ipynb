```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from permutations import finding_N
```

This does have a long run-time due to there being a lot of repeats in the finding_N function. If you want it to be faster you can adjust the number of repeats to be lower (in permutations.py). However, lowering the number of repeats does make the answer more variable.

```python
results = finding_N([0.5, 0.5], [0.9, 0.1]) 
#the input contains probability of no disease or disease without vaccine 
#the second input contains the probability of no disease or disease with the vaccine
```

```python
results
```

```python
probabilites = results[-1]
```

```python
#Plot of Probability of Success of Based on Sample Size
keys = list(probabilites.keys())
prob_vals = list(probabilites.values())

plt.plot(keys, prob_vals, label='Probability')

plt.xlabel('Number of Samples')
plt.ylabel('Probability')
plt.title('Probability of 90% Reaching Significance vs Number of Samples')

plt.show()
```
